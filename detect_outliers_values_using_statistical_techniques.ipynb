# Task
Generate a random dataset, visualize it using box, scatter, and histogram plots, detect outliers using statistical techniques, and visualize the detected outliers on the plots.

## Generate random data

### Subtask:
Create a random dataset for analysis.


**Reasoning**:
The first step is to import the numpy library and generate the random data as instructed.



```python
import numpy as np

data = np.random.normal(loc=50, scale=10, size=100)
print(data[:5])
```

## Visualize data

### Subtask:
Generate box, scatter, and histogram plots to visualize the data distribution.


**Reasoning**:
Generate a figure with three subplots for box, scatter, and histogram plots and plot the data on them.



```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(y=data, ax=axes[0])
axes[0].set_title("Box Plot of Data")

axes[1].scatter(range(len(data)), data)
axes[1].set_title("Scatter Plot of Data")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Data Value")


sns.histplot(data, ax=axes[2], kde=True)
axes[2].set_title("Histogram of Data")
axes[2].set_xlabel("Data Value")
axes[2].set_ylabel("Frequency")


plt.tight_layout()
plt.show()
```

## Detect outliers

### Subtask:
Use statistical techniques to identify outliers in the dataset.


**Reasoning**:
Calculate the IQR and determine the upper and lower bounds for outlier detection, then identify the outliers.



```python
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = data[(data < lower_bound) | (data > upper_bound)]
print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

## Visualize outliers

### Subtask:
Visualize the identified outliers on the plots.


**Reasoning**:
Regenerate the plots and visualize the outliers on the scatter plot.



```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(y=data, ax=axes[0])
axes[0].set_title("Box Plot of Data")

axes[1].scatter(range(len(data)), data, label='Data')
axes[1].scatter(np.where((data < lower_bound) | (data > upper_bound))[0], outliers, color='red', marker='o', label='Outliers')
axes[1].set_title("Scatter Plot of Data with Outliers")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Data Value")
axes[1].legend()


sns.histplot(data, ax=axes[2], kde=True)
axes[2].set_title("Histogram of Data")
axes[2].set_xlabel("Data Value")
axes[2].set_ylabel("Frequency")


plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   A random dataset of 100 numbers from a normal distribution with a mean of 50 and standard deviation of 10 was successfully generated.
*   The data distribution was visualized using box, scatter, and histogram plots.
*   Using the IQR method, the first quartile (Q1) was calculated as 43.45, the third quartile (Q3) as 55.65, and the IQR as 12.21.
*   The lower bound for outlier detection was determined to be 25.14 and the upper bound to be 73.96.
*   One outlier was identified at the value of 21.66.
*   The identified outlier was successfully visualized on the scatter plot using a distinct color and marker.

### Insights or Next Steps
