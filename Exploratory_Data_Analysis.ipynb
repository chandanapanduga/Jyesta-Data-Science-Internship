## Load data

### Subtask:
Load a random dataset using a library like Seaborn.


**Reasoning**:
Load the 'tips' dataset from seaborn and display its head, column names, and data types.



```python
import seaborn as sns
import pandas as pd

# Load the 'tips' dataset
df = sns.load_dataset('tips')

# Display the first few rows
display(df.head())

# Print column names and their data types
print(df.info())
```

## Visualize data using seaborn

### Subtask:
Create box, scatter, and heat map plots using Seaborn to explore the dataset.


**Reasoning**:
Create the requested box plot, scatter plot, and heatmap using Seaborn to visualize the data.



```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a box plot to visualize the distribution of total_bill across different day categories.
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Distribution of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount ($)')
plt.show()

# 2. Generate a scatter plot to examine the relationship between total_bill and tip.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Relationship between Total Bill and Tip')
plt.xlabel('Total Bill Amount ($)')
plt.ylabel('Tip Amount ($)')
plt.show()

# 3. Compute the correlation matrix for the numerical columns of the DataFrame (total_bill, tip, and size).
numerical_cols = ['total_bill', 'tip', 'size']
correlation_matrix = df[numerical_cols].corr()

# 4. Create a heatmap of the correlation matrix calculated in the previous step.
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
